# Brute-force

```python
class Solution:
    def isRectangleCover(self, rectangles):
        
        # Find min/max
        rectangles.sort()
        x_min, x_max = float('inf'), -float('inf')
        y_min, y_max = float('inf'), -float('inf')
                
        for x0, y0, x1, y1 in rectangles:
            if x0 < x_min: x_min = x0
            if y0 < y_min: y_min = y0
            if x1 > x_max: x_max = x1
            if y1 > y_max: y_max = y1
        
        # Check every point
        for x in range(x_min, x_max):
            for y in range(y_min, y_max):
                flag = False
                for x0, y0, x1, y1 in rectangles: 
                    if x0 < x+0.5 < x1 and y0 < y+0.5 < y1:
                        if flag:
                            return False
                        else:
                            flag = True
                if not flag:
                    return False
        
        return True
```

```python
rectangles = [
    [1,3,2,4],
    [1,1,3,3],
    [3,1,4,2],
    [3,2,4,4],
    [2,3,3,4]
]

Solution().isRectangleCover(rectangles)
```

# Check Side by Side

```python
class Solution:
    
    def isRectangleCover(self, rectangles):
        
        # Find min/max
        x_min, x_max = float('inf'), -float('inf')
        y_min, y_max = float('inf'), -float('inf')
                
        for x0, y0, x1, y1 in rectangles:
            if x0 < x_min: x_min = x0
            if y0 < y_min: y_min = y0
            if x1 > x_max: x_max = x1
            if y1 > y_max: y_max = y1
                
                
        # Check intersections
        n = len(rectangles)
        
        for i in range(n):
            for j in range(i+1, n):
                
                if rectangles[i] == rectangles[j]: return False
                
                XA, YA, XB, YB = rectangles[i]
                xA, yA, xB, yB = rectangles[j]
                
                Xflag = (xA <= XA < xB <= XB)
                xflag = (XA <= xA < XB <= xB)
                Yflag = (yA <= YA < yB <= YB)
                yflag = (YA <= yA < YB <= yB)
                
                if Xflag and Yflag: return False
                if xflag and Yflag: return False
                if Xflag and yflag: return False
                if xflag and yflag: return False
                
        
        # Check left/right sides
        rectangles.sort(key=lambda r: r[1])
        
        for i, r in enumerate(rectangles):
            print(i, r)
        print()
        
        for i in range(n):
            XA, YA, XB, YB = rectangles[i]
            
            for orient in ['left', 'right']:
                X = XA if orient == 'left' else XB
                
                if x_min < X < x_max:
                    for j in range(n):
                        xA, yA, xB, yB = rectangles[j]
                        x = xA if orient == 'right' else xB
                        
                        print(orient)
                        print(i, '  X:', X, '  Y:', YA, YB)
                        print(j, '  x:', x, '  y:', yA, yB)
                        print()
                            
                        if x < X:
                            continue
                        elif x == X: 
                            if yA <= YA and YA <= yB:
                                print('ok!')
                                print()
                                if yB < YB:
                                    # COVER SIDE TO POINT yB
                                    YA = yB
                                else:
                                    # SIDE IS COVERED!
                                    break 
                            if YA < yA:
                                continue
                        elif X < x:
                            continue
                    else:
                        # not covered after all
                        return False
        
        # Check top/bottom sides
        rectangles.sort(key=lambda r: r[0])
        
        for i in range(n):
            XA, YA, XB, YB = rectangles[i]
            
            for orient in ['top', 'bottom']:
                Y = YA if orient == 'top' else YB
                
                if y_min < Y < y_max:
                    for j in range(n):
                        xA, yA, xB, yB = rectangles[j]
                        y = yA if orient == 'bottom' else yB
                        
                        print(orient)
                        print(i, '  X:', XA, XB, '  Y:', Y)
                        print(j, '  x:', xA, xB, '  y:', y)
                        print()
                            
                        if y < Y:
                            continue
                        elif y == Y: 
                            if xA <= XA and XA <= xB:
                                print('ok!')
                                print()
                                if xB < XB:
                                    # COVER SIDE TO POINT xB
                                    XA = xB
                                else:
                                    # SIDE IS COVERED!
                                    break 
                            if XA < xA:
                                continue
                        elif Y < y:
                            continue
                    else:
                        # not covered after all
                        return False

        return True
```

# Calculate Area

```python
class Solution:
    
    def isRectangleCover(self, rectangles):
        
        # Find min/max
        x_min, x_max = float('inf'), -float('inf')
        y_min, y_max = float('inf'), -float('inf')
                
        for x0, y0, x1, y1 in rectangles:
            if x0 < x_min: x_min = x0
            if y0 < y_min: y_min = y0
            if x1 > x_max: x_max = x1
            if y1 > y_max: y_max = y1
                
        # Check intersections
        n = len(rectangles)
        
        for i in range(n):
            for j in range(i+1, n):
                
                if rectangles[i] == rectangles[j]: return False
                
                XA, YA, XB, YB = rectangles[i]
                xA, yA, xB, yB = rectangles[j]
                
                Xflag = (xA <= XA < xB)
                xflag = (XA <= xA < XB)
                Yflag = (yA <= YA < yB)
                yflag = (YA <= yA < YB)
                
                if Xflag and Yflag: return False
                if xflag and Yflag: return False
                if Xflag and yflag: return False
                if xflag and yflag: return False
                
        
        # Calc Area
        area_calc = (x_max-x_min)*(y_max-y_min)
        area_meas = 0
        
        for x0, y0, x1, y1 in rectangles:
            area_meas += (x1-x0)*(y1-y0)

        return area_meas == area_calc
```

# Corners + Area

```python
class Solution:
    
    def isRectangleCover(self, rectangles):
        
        # Find min/max
        x_min, x_max = float('inf'), -float('inf')
        y_min, y_max = float('inf'), -float('inf')
                
        for x0, y0, x1, y1 in rectangles:
            if x0 < x_min: x_min = x0
            if y0 < y_min: y_min = y0
            if x1 > x_max: x_max = x1
            if y1 > y_max: y_max = y1
                
        # Check if each Corner has its pair
        corners = set()
        for x0, y0, x1, y1 in rectangles:
            corners ^= {(x0,y0), (x1,y1), (x0,y1), (x1,y0)}
        
        if corners != {(x_min, y_min), 
                       (x_max, y_max), 
                       (x_min, y_max), 
                       (x_max, y_min)}:
            return False
        
        # Calc Area
        area_calc = (x_max-x_min)*(y_max-y_min)
        area_meas = 0
        
        for x0, y0, x1, y1 in rectangles:
            area_meas += (x1-x0)*(y1-y0)
        
        if area_calc != area_meas:
            return False
        
        # If everything OK
        return True
```

# Optimization

```python
class Solution:
    
    def isRectangleCover(self, rectangles):
        
        x0, x1, y0, y1 = None, None, None, None
        
        corners = set()
        area = 0
                
        for xA, yA, xB, yB in rectangles:
            
            corners ^= {(xA,yA), (xB,yB), (xA,yB), (xB,yA)}
            area += (xB-xA)*(yB-yA)
            
            if x0 is None or xA <= x0 and yA <= y0: x0, y0 = xA, yA
            if x1 is None or xB >= x1 and yB >= y1: x1, y1 = xB, yB
        
        return (corners == {(x0, y0), (x1, y1), (x0, y1), (x1, y0)} and \
                area == (x1-x0)*(y1-y0))
```
